# 基本爬蟲

```python
import requests
from bs4 import BeautifulSoup as bs # 用 bs 代替 BeautifulSoup

url = 'https://www.ptt.cc/bbs/MobileComm/index3965.html'
res = requests.get(url) # 向 url 提出請求
print res.status_code
print '代碼意義請參考: https://zh.wikipedia.org/wiki/HTTP%E7%8A%B6%E6%80%81%E7%A0%81'
```

```python
print res.text # 回應的 html 內容
```

```python
soup = bs(res.text,'html') # 使用 BeautifulSoup 做剖析
titles = soup.select('.r-ent a') # css 選擇器, 自行研究該網頁標籤, 注意回傳值為 list 
for title in titles:
    print title
    #print title.text # 去掉標籤的文字內容
    #print title['href'] # 取出標籤的 href 屬性下帶的值
```

# 存入檔案

```python
path_output = '/Users/Yehkuangting/Desktop/pttMobile.txt'

with open(path_output, 'w') as f: # 開啟檔案, 寫入模式
    for title in titles:
        print title.text
        # f.write(title.text) # 此行錯誤, 須將爬下來的 unicode 型態轉成 str
        f.write(title.text.encode('utf8') + '\n')
```

# 編碼
