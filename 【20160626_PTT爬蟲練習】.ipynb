# PTT爬蟲練習 Tai-Travel

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://www.ptt.cc/bbs/Tai-travel/index.html')
print res.text
```

```python
soup = BeautifulSoup(res.text,'html5lib')
res.encoding='utf8'
ele = soup.select('div.title a')

pttlist = []
for one in ele:
    pttlist.append(one.text)
    print one.text
# print(pttlist)
# print(pttlist)
```

# 將爬到的資料寫入指定檔案中

```python

fid = open("C:\Users\BIG DATA\Desktop\ptt.txt","w")

for title in pttlist:
    fid.write(title.encode('utf8'))
    print(type(title.encode('utf8')))
fid.close()
```
